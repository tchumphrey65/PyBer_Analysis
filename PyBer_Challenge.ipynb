### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
#%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics
import scipy.stats as sts
import datetime

# File to Load (Remember to change these)
city_data_to_load = "/Users/Tim/Documents/OSU_Bootcamp/Challenge_5/Resources/city_data.csv"
ride_data_to_load = "/Users/Tim/Documents/OSU_Bootcamp/Challenge_5/Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

# Get the number of rides for urban cities.
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
print(total_rides_by_city_type)



```

```python
# 2. Get the total drivers for each city type

total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"] 
print(total_drivers_by_city_type)

```

```python
#  3. Get the total amount of fares for each city type

total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"] 
print(total_fares_by_city_type)
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride_city_type = total_fares_by_city_type/total_rides_by_city_type
print(avg_fare_per_ride_city_type)
```

```python
# 5. Get the average fare per driver for each city type. 

avg_fare_per_driver_city_type = total_fares_by_city_type/total_drivers_by_city_type
print(avg_fare_per_driver_city_type)
```

```python
#  6. Create a PyBer summary DataFrame. 
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame()
pyber_summary_df["Total Rides"] = total_rides_by_city_type
pyber_summary_df["Total Drivers"] = total_drivers_by_city_type
pyber_summary_df["Total Fares"] = total_fares_by_city_type
pyber_summary_df["Avg Fare per Ride"] = avg_fare_per_ride_city_type
pyber_summary_df["Avg Fare per Driver"] = avg_fare_per_driver_city_type

#  8. Format the columns.


pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:}".format)
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Avg Fare per Ride"] = pyber_summary_df["Avg Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Avg Fare per Driver"] = pyber_summary_df["Avg Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head(10)
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

total_fares_by_date = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
total_fares_by_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fares_by_date = total_fares_by_date.reset_index()

total_fares_by_date.head(10)


```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_table = pd.pivot_table(total_fares_by_date, index = "date", columns = "type", values = "fare")
pivot_table.head(10)

```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pivot_fare_dates_df = pivot_table.loc['2019-01-01':'2019-04-29']
pivot_fare_dates_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

#type(pivot_fare_dates_df.index)

type(pivot_fare_dates_df.index)

pivot_fare_dates_df.index = pd.to_datetime(pivot_fare_dates_df.index)
pivot_fare_dates_df.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_fare_dates_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pivot_fare_dates_df = pivot_fare_dates_df.resample('W').sum()
pivot_fare_dates_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
pivot_fare_dates_df.plot(figsize=(18,8))
plt.ylim(0,3000)
plt.xlabel('Month')
plt.ylabel('Fare ($)')
plt.title('Total Fare by City Type');
plt.legend(loc='upper right')
plt.savefig("/Users/Tim/Documents/OSU_Bootcamp/Challenge_5/Analysis/Fare_by_city_chart.png")
plt.show()
```
